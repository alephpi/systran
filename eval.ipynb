```python
from data import load_vocabulary
import torch
import torch.nn as nn

tgt_vocab = '../corpus/vocab_en_fr.txt'
target_vocabulary, target_vocabulary_rev = load_vocabulary(tgt_vocab)

mask = torch.load('./penalty_mask.pt')
```

```python
acc_vocab = [target_vocabulary_rev[i]
             for i in mask.nonzero().squeeze().tolist()]
```

```python
def analyse(dt: dict[str, int], dr: dict[str, int], dh: dict[str, int], acc_vocab: list[str]):
    """
    t: repetition words in train set
    r: repetition words in references
    h: repetition words in hypothesis
    """
    t = set(dt.keys())
    r = set(dr.keys())
    h = set(dh.keys())
    # words that avoids repeating
    avoid = r.difference(h)
    new = h.difference(r)
    remain = h
    fail1 = avoid.difference(t)
    # rep words that don't repeat in train
    fail2 = h.difference(t)
    # rep words repeat in train
    fail3 = h.intersection(t)
    d1 = {k: dt[k] for k in fail1}
    d2 = {k: dt[k] for k in fail2}
    d3 = {k: dt[k] for k in fail3 if k in acc_vocab}
    a = {k: dt[k] for k in avoid if k in acc_vocab}
    return d1, d2, d3, a
```

```python
d1, d2, d3, a = analyse(rep0_dico, rep_ref_dico, rep6_dico, acc_vocab)
```

```python
import os
import subprocess
import json
from wordrep.utils import BaseRepDetector


def eval_score(directory='./corpus/', ref_path='../corpus/rep_test.fr'):
    c = "sacrebleu"  # Replace with your desired Bash command
    rep_detector = BaseRepDetector('fr')
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            # Process the file
            if filename.endswith('fr'):
                path = os.path.join(directory, filename)
                command = f'{c} {path} -i {ref_path}'
                # Execute the command
                process = subprocess.Popen(
                    command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
                # Wait for the command to finish and capture the output
                output, _ = process.communicate()
                # Decode the output
                output = output.decode('utf-8')
                o = json.loads(output)
                reps = rep_detector.detect_corpus(path, vis=False)
                print(filename, f"bleu={o['score']}, rep={len(reps)}")
```

```python
eval_score(directory='./corpus/train-approach')
```
